# Data Visualization For California Hosuing Dataset

```python
import pandas as pd
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn import preprocessing
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
```

```python
housing_df = pd.read_csv("https://raw.githubusercontent.com/nyangweso-rodgers/Data-Analytics/refs/heads/main/sample-datasets/02-housing.csv", usecols = ['longitude', 'latitude', 'median_house_value'])
print("Housing Dataset Shape: ", housing_df.shape)
housing_df.head()
```

## Visualize the Data

### Heatmap
- Create a heatmap based on the median price in a block

```python
sns.scatterplot(data = housing_df, x = 'longitude', y = 'latitude', hue = 'median_house_value')
```

__Remarks__
1. We see that most of the expensive houses are on the west coast of California with different areas that have clusters of moderately priced houses. This is expected as typically waterfront properties are worth more than houses that are not on the coast.
