# Introduction to Matplotlib

```python
# import the packages
import matplotlib.pyplot as plt
import numpy as np
```

## Plot Single Graph
* When you want to draw a single graph, you can use __plt.plot()__ directly, but using __plt.subplots()__ is also OK.
* we can give two arguments to subplots functions: nrows and ncols. If given in that order, we don't need to type the arg names

## Matplotlib Subplots

* The __matplotlib.pyplot.subplots__ method provides a way to plot multiple plots on a single figure. Given the number of __rows__ and __columns__, it returns a tuple (__fig__, __ax__), giving a single figure __fig__ with an array of axes __ax__.
  
* __subplots()__ without arguments returns a Figure and a single Axes, which we can unpack using the syntax bellow.
  
* __Parameters__:
  1. __nrows__, __ncols__: Number of rows and columns of the subplot grid. Both of these are optional with a default value of 1.
  2. __sharex__, __sharey__: Specifies sharing of properties between axes. Possible values are none, all, row, col or a boolean with a default value of False.
  3. __figsize__ set the total dimension of our figure

```python

# Subplots without arguments
fig, ax = plt.subplots(figsize=(12,4))
fig.suptitle('A single ax with no data')
```

```python
#plt.style.use('seaborn-whitegrid') # you can switch to classic
# 1 row, 3 columns
plt.subplots(nrows=1, ncols=3, sharex = False, sharey = False, squeeze = True, figsize=(16,5))
```

Alternatively,

```python
# using fig, axes
fig, axes = plt.subplots(1, 3, figsize=(18, 4), sharex=True)
fig.suptitle('1 row x 3 columns axes with no data')
# set the title for the first chart
axes[0].set_title('Title of the first chart')
```

```python
fig, ax = plt.subplots(2, 2, figsize=(16,4))
x = np.linspace(0, 8, 1000)

ax[0, 0].plot(x, np.sin(x), 'g') #row=0, col=0
ax[1, 0].plot(x, np.tan(x), 'k') #row=1, col=0
ax[0, 1].plot(range(100), 'b') #row=0, col=1
ax[1, 1].plot(x, np.cos(x), 'r') #row=1, col=1
#fig.show()
```

## Plot using Function

```python
# 2x2, a total of 4 graphs
n_rows = 2
n_cols = 2

# Create a figure and subplots of size 2x2
fig, axes = plt.subplots(nrows=n_rows, ncols=n_cols, figsize=(18,6))  

for row_num in range(n_rows):
    for col_num in range(n_cols):
        ax = axes[row_num][col_num]
        ax.plot(np.random.rand(10))
        
fig.suptitle("Main Title")
fig.tight_layout()  # Smaller padding
plt.show()
```
